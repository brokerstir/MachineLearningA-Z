```
from google.colab import files
uploaded = files.upload()
```

```
import pandas as pd
import io

# read csv
train_df = pd.read_csv(io.StringIO(uploaded['train.csv'].decode('utf-8')))

# safety: limit number of rows to display
pd.set_option("display.max_rows", 10)

# look at data
train_df.head()
```

```
# Neighborhoods to the number of homes sold in each neighborhood
train_df['Neighborhood'].value_counts().head(10).plot.bar()
```

```
# change figzie (width, height) and color
train_df['Neighborhood'].value_counts().head(10).plot.bar(figsize=(12, 6), color='mediumvioletred')
```

```
# alter figsize, color, and fontsize
(train_df['Neighborhood'].value_counts().head(10) / len(train_df)).plot.bar(
    figsize=(12, 6), 
    color='mediumvioletred', 
    fontsize=16, 
    title='Proportions of Neighborhoods',)
```

```
# plot using matplot lib
import matplotlib.pyplot as plt

ax = train_df['Neighborhood'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
ax.set_title("Proportions of Neighborhoods", fontsize=20)
```

```
import seaborn as sns

ax = train_df['Neighborhood'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)

ax.set_title("Proportions of Neighborhoods", fontsize=20)
sns.despine(bottom=True, left=True)
```

```
# subplots takes two arguments. The first one controls the number of rows, the second one the number of columns.
fig, axarr = plt.subplots(2, 1, figsize=(12, 8))

train_df['OverallQual'].value_counts().sort_index().plot.bar(ax=axarr[0])

train_df['OverallCond'].value_counts().sort_index().plot.bar(ax=axarr[1])
```
